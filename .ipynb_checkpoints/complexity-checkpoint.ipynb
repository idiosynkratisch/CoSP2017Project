```python
ndepth = set()
nwidth = set()
nbalanced = set()

for conv in corpus.iter_transcripts(display_progress=False):
    for utt in conv.utterances:
        for tree in utt.trees:
            ndepth.add((conv.conversation_no, utt.transcript_index,
                        complexity.ndepth(tree), tree, utt.text))
            nwidth.add((conv.conversation_no, utt.transcript_index,
                        complexity.nwidth(tree), tree, utt.text))
            nbalanced.add((conv.conversation_no, utt.transcript_index,
                           complexity.nbalanced(tree), tree, utt.text))

ndepth = list(ndepth)
nwidth = list(nwidth)
nbalanced = list(nbalanced)
            
ndepth.sort(key=lambda x: x[2])
nwidth.sort(key=lambda x: x[2])
nbalanced.sort(key=lambda x: x[2])

print 'Least complex trees by normalized depth:'
for i, item in enumerate(ndepth[:10], 1):
    print '{}.\t{}'.format(i, str(item[2]))
    
print 'Most complex trees by normalized depth:'
for i, item in enumerate(reversed(ndepth[len(ndepth)-10:]), 1):
    print '{}.\t{}'.format(i, str(item[2]))
    
print 'Least complex trees by normalized width:'
for i, item in enumerate(nwidth[:10], 1):
    print '{}.\t{}'.format(i, str(item[2]))
    
print 'Most complex trees by normalized width:'
for i, item in enumerate(reversed(nwidth[len(nwidth)-10:]), 1):
    print '{}.\t{}'.format(i, str(item[2]))

print 'Least complex trees by normalized depth*width:'
for i, item in enumerate(nbalanced[:10], 1):
    print '{}.\t{}'.format(i, str(item[2]))
    
print 'Most complex trees by normalized depth*width:'
for i, item in enumerate(reversed(nbalanced[len(nbalanced)-10:]), 1):
    print '{}.\t{}'.format(i, str(item[2]))
    
ndepth.sort(key=lambda x: abs(x[1] - 1))
nwidth.sort(key=lambda x: abs(x[1] - 1))
nbalanced.sort(key=lambda x: abs(x[1] - 1))

print 'Most average trees by normalized depth:'
for i, item in enumerate(ndepth[:10], 1):
    print '{}.\t{}'.format(i, str(item[2]))
    
print 'Most average trees by normalized width:'
for i, item in enumerate(nwidth[:10], 1):
    print '{}.\t{}'.format(i, str(item[2]))

print 'Most average trees by normalized depth*width:'
for i, item in enumerate(nbalanced[:10], 1):
    print '{}.\t{}'.format(i, str(item[2]))
```
